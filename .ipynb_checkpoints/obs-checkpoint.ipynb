```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
import cartopy.crs as ccrs
from matplotlib import ticker
import cartopy.feature as cfeature
```

```python
#Load the data 

CESM2 = xr.open_dataset("data/CESM2.nc") #  'ScenarioMIP.NCAR.CESM2.ssp585.Oyr.gn'
MRI_ESM2 = xr.open_dataset("data/MRI-ESM2-0.nc") #  'ScenarioMIP.MRI.MRI-ESM2-0.ssp585.Oyr.gn'
CMCC_ESM2 = xr.open_zarr("/Users/saituumi/miniconda3/e440/Project/loading_data/CMCC_ESM2.zarr", consolidated=True) #  'ScenarioMIP.CMCC.CMCC-ESM2.ssp585.Oyr.gn'
```

```python
CESM2_2100 = CESM2.sel(lev = 5.000000e+02, time=CESM2.time.values[-1]).mean(dim =[ "member_id","dcpp_init_year"])
CESM2_2015 = CESM2.sel(lev = 5.000000e+02, time=CESM2.time.values[0]).mean(dim =[ "member_id","dcpp_init_year"])
CESM2_diff = CESM2_2100 - CESM2_2015
```

```python
MRI_ESM2_2100 = MRI_ESM2.sel(lev = 1.0000e+00 , time=MRI_ESM_0.time.values[-1]).mean(dim =[ "member_id","dcpp_init_year"])
MRI_ESM2_2015 = MRI_ESM2.sel(lev = 1.0000e+00 , time=MRI_ESM_0.time.values[0]).mean(dim =[ "member_id","dcpp_init_year"])
MRI_ESM2_diff = MRI_ESM_0_2100  - MRI_ESM_0_2015
```

```python
CMCC_ESM2_2100 = CMCC_ESM2.sel(lev = CMCC_ESM2.lev.values[0], time=CMCC_ESM2.time.values[-1] ).mean(dim =[ "member_id","dcpp_init_year"])
CMCC_ESM2_2015 = CMCC_ESM2.sel(lev = CMCC_ESM2.lev.values[0], time=CMCC_ESM2.time.values[0] ).mean(dim =[ "member_id","dcpp_init_year"])
CMCC_ESM2_diff = CMCC_ESM2_2100 - CMCC_ESM2_2015
```

```python
plt.figure(figsize=(12, 6))

ax = plt.axes(projection=ccrs.PlateCarree())

# Plot the data
mesh = ax.pcolormesh(CESM2_diff.lon, CESM2_diff.lat, CESM2_diff.ph,  shading='auto', cmap='viridis')

# Add coastlines and grid
ax.coastlines()
ax.gridlines(draw_labels=True)

cbar = plt.colorbar(mesh,ax=ax, label='ph')
cbar.locator = ticker.AutoLocator()
cbar.set_ticks(cbar.locator.tick_values(-0.7, -0.2))

plt.title('Change of Ocean pH for CESM2 (2100 - 2015)')
plt.show()
```

```python
plt.figure(figsize=(12, 6))

ax = plt.axes(projection=ccrs.PlateCarree())

# Plot the data
mesh = ax.pcolormesh(CMCC_ESM2_diff.longitude, CMCC_ESM2_diff.latitude, CMCC_ESM2_diff.ph,  shading='auto', cmap='viridis')
 
# Add coastlines and grid
ax.coastlines()
ax.gridlines(draw_labels=True)

cbar = plt.colorbar(mesh,ax=ax, label='ph')
cbar.locator = ticker.AutoLocator()
cbar.set_ticks(cbar.locator.tick_values(-0.7, -0.2))

plt.title('Change of Ocean pH for CMCC-ESM2 (2100 - 2015)')
plt.show()
```
